### Задание 1. 
Spacy tagger. Популярные NER, персоны, организации

```python
 # Объединяем все слова в 1 список
def words_union(df):
    words_list = []
    for line in df['tweet_token_filtered']:
        #line_words = ' '.join(line)#.split()
        words_list.extend(line)
    words = words_list#' '.join(words_list)
    return words
```

```python
words_list = []
for line in combine_df['tweet_token_filtered']:
    words_list.extend(line)
    break
print(words_list)
```

```python
#!pip install nltk

#!pip install spacy==2.3.5

#!pip install https://github.com/explosion/spacy-models/releases/download/en_core_web_sm-2.3.1/en_core_web_sm-2.3.1.tar.gz

#!pip install pyresparser
```

```python
#!spacy download en_core_web_lg
#!pip install https://github.com/explosion/spacy-models/releases/download/en_core_web_md-2.3.1/en_core_web_md-2.3.1.tar.gz --no-deps
```

```python
words = words_union(combine_df)
max_word_limit_spacy = 10000
nlp = en_core_web_md.load()
doc = nlp(str(words[:max_word_limit_spacy]))
tokens_list = [(ent.text, ent.label_) for ent in doc.ents]
tokens_list
```

```python
# TOP-20 популярных NER
collections.Counter(tokens_list).most_common(20)
```

```python
# TOP-20 популярных личностей
person_tokens_list = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ == 'PERSON']
collections.Counter(person_tokens_list).most_common(20)
```

```python
# TOP-20 популярных организаций
org_tokens_list = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ == 'ORG']
collections.Counter(org_tokens_list).most_common(20)
```

```python
# сколько уникальных NER найдено с помощью spacy?
len(collections.Counter(tokens_list))
```

```python
words = words_union(combine_df)
max_word_limit_spacy = 10000
tokens_list = [(' '.join(c[0] for c in chunk), chunk.label() ) for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(str(words[:max_word_limit_spacy])))) if hasattr(chunk, 'label') ]
```

```python
tokens_list
```

```python
# TOP-20 популярных NER
collections.Counter(tokens_list).most_common(20)
```

```python
# TOP-20 популярных личностей
person_tokens_list = [(ent[0], ent[1]) for ent in tokens_list if ent[1] == 'PERSON']
collections.Counter(person_tokens_list).most_common(20)
```

```python
# TOP-20 популярных организаций
org_tokens_list = [(ent[0], ent[1]) for ent in tokens_list if ent[1] == 'ORGANIZATION']
collections.Counter(org_tokens_list).most_common(20)
```

```python
# сколько уникальных NER найдено с помощью spacy?
len(collections.Counter(tokens_list))
```
